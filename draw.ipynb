```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('test.csv')

ticks = data[data.type == 'tick']
jvm = data[data.type == 'jvm']
potential = data[data.type == 'potential']
plt.figure(figsize=(30, 15))
plt.plot(ticks['time_point'], ticks['size'])

jvm['size'].fillna(0)
potential['size'].fillna(0)

for i in jvm:
    plt.scatter(i['time_point'], i['size'], color='red')

for i in potential:
    plt.scatter(i['time_point'], i['size'], color='blue')

plt.show()
```
